```python
import matplotlib.pyplot as plt
demo=data.groupby("year")["amount"].sum()
demo
plt.figure(figsize=(10,6))
plt.plot(demo.index,demo.values,marker="o",markerfacecolor="red",color="black",markeredgecolor="green",linestyle="--",linewidth=1.5,label="Amount")
plt.title("Year wise Amount investment", color="red",fontsize=16)
plt.grid(linestyle="--",color="skyblue")
plt.xlabel("year",color="blue",fontsize=12)
plt.ylabel("Total amount",color="blue",fontsize=12)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
Demo=data["city"].value_counts().head(10)
Demo

color=["blue","black","red","yellow","pink","green","purple","orange","brown","skyblue"]

plt.figure(figsize=(10,6))
plt.bar(Demo.index,Demo.values,color=color,alpha=0.4,edgecolor="black")
plt.xticks(rotation=45)
plt.title("Top 10 startup company's in India")
plt.xlabel("City Names")
plt.ylabel("Number of startup")
plt.grid(True)

plt.show()
```

```python
from matplotlib.lines import lineStyles
demo=data["vertical"].value_counts().head(10)
demo

color=["blue","black","red","yellow","pink","green","purple","orange","brown","skyblue"]
plt.figure(figsize=(10,6))

plt.barh(demo.index[::-1],demo.values[::-1],color=color,alpha=0.6,edgecolor="black")
plt.title("Top 10 Sectors in Startup in India",fontsize=16)
plt.xlabel("Number of Startup comany")
plt.ylabel("Sectors")
plt.grid(axis= 'x' ,linestyle="--",color="grey")

plt.show()
```

```python
demo=data["round"].value_counts().head(7)
demo
color=["blue","black","red","yellow","pink","green","purple"]
plt.figure(figsize=(10,6))

plt.pie(demo.values,labels=demo.index,autopct="%1.1f%%",colors=color)
plt.title("Rounds Distribution in Percentage")

plt.show()
```

```python
#how are startup funding amounts distribution?
demo=data["amount"].value_counts().head(7)
demo
plt.figure(figsize=(10,6))
plt.hist(data["amount"],bins=10,edgecolor="black")
plt.title("Distribution of startup funding")


```

```python
#relationship between funding year and funding amount?
demo=data["year"].value_counts().head(7)
demo

plt.figure(figsize=(10,6))
plt.scatter(data["year"],data["amount"],color="skyblue")
plt.title("Funding amount by year")
plt.xlabel("year")
plt.ylabel("amount")
plt.grid(linestyle="--")
plt.show()
```

```python
#which funding round occurs most frequently?
import seaborn as sns

demo=data["round"].value_counts().head(10).index
plt.figure(figsize=(10,6))
sns.countplot(data,y="round",order=demo,palette="viridis")
plt.title("Most frequently funding round")
```
